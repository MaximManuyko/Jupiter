**Шаг 4.** Напишите функцию get_mean_monday(clicks_values: np.ndarray), которая возвращает среднее значение кликов среди всех точек только в разрезе понедельников. Постарайтесь использовать срезы данных.

```python
import numpy as np

clicks_values = [
    [319, 265, 319, 328],
    [292, 274, 292, 301],
    [283, 301, 274, 283],
    [328, 364, 328, 319],
    [391, 355, 373, 337],
    [445, 418, 409, 445],
    [481, 400, 481, 409],
    [333, 267, 333, 344],
    [300, 278, 300, 311],
    [289, 311, 278, 289],
    [344, 388, 344, 333],
    [421, 377, 399, 355],
    [487, 454, 443, 487],
    [531, 432, 531, 443],
    [312, 264, 312, 320],
    [288, 272, 288, 296],
    [280, 296, 272, 280],
    [320, 352, 320, 312],
    [376, 344, 360, 328],
    [424, 400, 392, 424],
    [456, 384, 456, 392],
    [347, 269, 347, 360],
    [308, 282, 308, 321],
    [295, 321, 282, 295],
    [360, 412, 360, 347],
    [451, 399, 425, 373],
    [529, 490, 477, 529],
    [581, 464, 581, 477]
]

clicks_values = np.array(clicks_values)

def get_mean_monday_all_points(clicks_values: np.ndarray):
    # Используем срез [0::7] для выбора данных по всем точкам и всем понедельникам
    selected_data = clicks_values[0::7]
    # Вычисляем среднее значение для выбранных данных
    mean_value = np.mean(selected_data)
    return mean_value

# Вызываем функцию и выводим результат
mean_value = get_mean_monday_all_points(clicks_values)
print("Среднее значение кликов по всем точкам и всем понедельникам:", mean_value)
```

**Шаг 5.** Напишите функцию get_max_click(clicks_values: np.ndarray, col_number: int), которая возвращает максимальное значение кликов за весь период для площадки (столбца) под номером col_number. Постарайтесь использовать срезы данных.

```python
import numpy as np

clicks_values = [
    [319, 265, 319, 328],
    [292, 274, 292, 301],
    [283, 301, 274, 283],
    [328, 364, 328, 319],
    [391, 355, 373, 337],
    [445, 418, 409, 445],
    [481, 400, 481, 409],
    [333, 267, 333, 344],
    [300, 278, 300, 311],
    [289, 311, 278, 289],
    [344, 388, 344, 333],
    [421, 377, 399, 355],
    [487, 454, 443, 487],
    [531, 432, 531, 443],
    [312, 264, 312, 320],
    [288, 272, 288, 296],
    [280, 296, 272, 280],
    [320, 352, 320, 312],
    [376, 344, 360, 328],
    [424, 400, 392, 424],
    [456, 384, 456, 392],
    [347, 269, 347, 360],
    [308, 282, 308, 321],
    [295, 321, 282, 295],
    [360, 412, 360, 347],
    [451, 399, 425, 373],
    [529, 490, 477, 529],
    [581, 464, 581, 477]
]

clicks_values = np.array(clicks_values)

def get_max_click(clicks_values: np.ndarray, col_number: int):
    col_data = clicks_values[:, col_number]
    max_click = np.max(col_data)
    return max_click

# Пройдемся циклом по всем столбцам и выведем максимальное значение для каждого
for col_number in range(clicks_values.shape[1]):
    max_click_value = get_max_click(clicks_values, col_number)
    print(f"Максимальное значение в столбце {col_number}: {max_click_value}")
```
